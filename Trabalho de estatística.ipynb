<h2>Funçoes de apoio</h2>


```python
def media(Mat,tam):
    lst = []
    for cont in range(tam):
        lst.append(np.mean(Mat[cont]))        
    #display(lst)
    return lst
```

```python

def grafico(lst,bin_edges):
    fig, ax = plt.subplots()
    ax.hist(lst, bin_edges, cumulative=False, density=True)
    sigma = np.std(lst)
    #média = 0 e desvio-padrão = 1
    mu = np.mean(lst)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma))
    ax.set_xlabel('x')
    ax.set_ylabel('Frequencia')
    plt.show
    return plt
```

```python
def MAT(elem, tam, tipo):
    np.random.seed(seed=0)
    X0 = []
    for y in range(50):
        if(tipo==0):
            I = [np.random.randn(tam)]
        if(tipo==1):
            I = [np.random.random(tam)]
        if(tipo==2):
            I = [np.random.randint(2,tam)]            
        if(tipo==3):
            I = [np.random.rand(tam)]               
        if(tipo==4):
            I = [np.random.random_sample(tam,)]                
        print(I)
        X0.append(I)
    return X0
```

<h2>1) Gerando 50 amostras de 5 aleatórias</2>

```python
X = MAT(50,5,0)
```

<h2>2 - Gerando a média das 50 amostras geradas</h2>

```python
Lista = []
Lista = media(X,50)
display(Lista)

```

<h2>3 -Crie um Histograma<h2>

```python
hist, bin_edges = np.histogram(Lista, bins=7)
```

<h2>4 -Plote a curva da distribuição sobre o histograma</h2>
